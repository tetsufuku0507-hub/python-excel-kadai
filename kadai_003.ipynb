```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from google.colab import drive

# Google Drive をマウント
drive.mount('/content/drive')

BASE_PATH = '/content/drive/MyDrive/年間売上表/'

# ① 複数ファイルからデータフレームを作成・連結
df_2022 = pd.read_excel(BASE_PATH + '2022_年間売上表.xlsx', sheet_name='Sheet1')
df_2023 = pd.read_excel(BASE_PATH + '2023_年間売上表.xlsx', sheet_name='Sheet1')

df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# 「商品」「売上年」ごとにグループ化し、合計金額に集約
df_summary = (
    df_all
    .groupby(['商品', '売上年'], as_index=False)['金額（千円）']
    .sum()
    .sort_values(['商品', '売上年'])
    .reset_index(drop=True)
)

print(df_summary)

# ② 新規 Excel ファイル「売上集計表.xlsx」に書き込み
OUTPUT_PATH = BASE_PATH + '売上集計表.xlsx'

df_summary.to_excel(OUTPUT_PATH, index=False, sheet_name='Sheet1')

# ヘッダー行（1行目）を薄いグレー（#F2F2F2）に設定
wb = load_workbook(OUTPUT_PATH)
ws = wb.active

header_fill = PatternFill(fill_type='solid', fgColor='F2F2F2')
for cell in ws[1]:
    cell.fill = header_fill

wb.save(OUTPUT_PATH)
print('売上集計表.xlsx を保存しました。')
```
